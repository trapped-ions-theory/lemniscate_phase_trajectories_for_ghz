```python
t_gate = 2*np.pi
t_arr = np.linspace(0, t_gate, 10001)

N_cutoff = 100
eta = 0.0
n_loops = 8
f_max = math.sqrt(2*n_loops)/2
f_arr = f_max*np.where(t_arr < t_arr[-1]/2, 
                                      np.exp(-2j*n_loops*t_arr),
                                     -np.exp(-2j*n_loops*t_arr))
psi0 = fock(N_cutoff, 0)
mx = 2
ham = ion_ham(f_arr, t_arr, eta, mx, N_cutoff)
result = sesolve(ham, psi0, t_arr, options={'store_final_state': True},
                e_ops=[position(N_cutoff), momentum(N_cutoff), num(N_cutoff)])
```
