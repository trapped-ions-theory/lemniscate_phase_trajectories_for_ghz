```python
from tdse_solutions import lemniscate_f
```

```python
t = np.linspace(0, 2*np.pi, 2001)
a = 0.7274788716591838
Omega_max = 0.95778915
x = Omega_max*(1 - np.cos(t))
y = Omega_max*np.sin(t)*(1 - a + a*np.cos(t))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

ax2.set_title('Lemniscate phase trajectory')
ax2.plot(x, y, color='g')
ax2.set_xlabel(r'$\operatorname{Re}\alpha$')
ax2.set_ylabel(r'$\operatorname{Im}\alpha$')
ax2.set_aspect('equal')


f = lemniscate_f(t, a)
ax1.set_title('Lemniscate pulse')
ax1.plot(t/2/np.pi, f.real, label=r'$\operatorname{Re} \Omega/\Omega_{max}$')
ax1.plot(t/2/np.pi, f.imag, label=r'$\operatorname{Im} \Omega/\Omega_{max}$')
ax1.set_xlabel('$t/t_{gate}$')
ax1.legend()
ax1.margins(x=0)

#plt.savefig('../figures/lemniscate_pulse.pdf')
plt.show()
```
